```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
jobs = pd.read_csv('local-file.csv')
jobs.head()
```

```python
print(jobs.nunique())
print(jobs.info())
```

```python
import numpy as np

# Convert 'nan' strings to actual NaN
jobs['min_years_experience'] = jobs['min_years_experience'].replace('nan', np.nan)

# Convert to float
jobs['min_years_experience'] = jobs['min_years_experience'].astype(float)

# Drop actual NaN values
jobs_clean = jobs['min_years_experience'].dropna()

# Calculate mode
if not jobs_clean.empty:
    mode_yrs_exp = jobs_clean.mode().iloc[0]
else:
    mode_yrs_exp = None  # Handle case where no valid mode exists

print(mode_yrs_exp)
```

```python
jobs['min_years_experience'].fillna('0', inplace=True)
```

```python
#Dropping columns becuase too many null values or no values in table
jobs.drop(columns=['region', 'zipcode', 'city', 'state_code', 'county', 'rural_type'], axis=1, inplace=True)
jobs.info()
```

```python
categories = ['career_area_name', 'company_name', 'employment_type_name', 'industry_clean', 'min_edulevel_name', 'remote_type_name']
numeric = ['max_salary', 'min_salary', 'min_years_experience', 'salary']
binary = ['isPartner', 'is_earn_and_learn', 'is_estimated_salary', 'is_gateway_job']
date_time = ['createdAt', 'expiredAt', 'updatedAt']

jobs[categories] = jobs[categories].astype('category')
jobs[binary] = jobs[binary].astype('int32')
jobs[numeric] = jobs[numeric].astype('float16')

for col in jobs[categories]:
    # print(f'Column: {col} has {jobs[col].nunique()} values')
    if jobs[col].nunique() < 40:
        print(f'{jobs[col].value_counts(dropna=False)}')
        print('='*110)
```

```python
jobs['employment_type_name'] = jobs['employment_type_name'].str.lower().replace(r'^full.*', 'full_time', regex=True)
jobs['employment_type_name'] = jobs['employment_type_name'].str.lower().replace(r'^part.*', 'part_time', regex=True)

print(jobs.employment_type_name.value_counts(dropna=False))

# jobs.info()
```

```python
jobs.rename(columns={'employment_type_name':'isPartTime'}, inplace=True)
```

```python
jobs.columns
```

```python
jobs['isPartTime'] = jobs['isPartTime'].apply(lambda x: 1 if x == 'part_time' else 0)
print(jobs['isPartTime'].value_counts())
```

```python
new_education = ["No Education Listed", "High school or GED", "Associate degree", "Bachelor's degree"]
jobs['min_edulevel_name'] = jobs['min_edulevel_name'].astype('category')
jobs['min_edulevel_name'] = jobs['min_edulevel_name'].cat.reorder_categories(new_education, ordered=True)
print(jobs['min_edulevel_name'].value_counts(dropna=False).reindex(new_education, fill_value=0))
```

```python
jobs.head()
```

```python
jobs[date_time] = jobs[date_time].apply(pd.to_datetime)
print(jobs[date_time].dtypes)
```

```python
jobs.to_csv('cleaned_jobs.csv', index=False)

from google.colab import files
files.download('cleaned_jobs.csv')
```
